```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
data1=pd.read_csv('Dataset1.csv')
data2=pd.read_csv('Dataset2.csv')

# run elbow method
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data2)
    distortions.append(kmeanModel.inertia_)
plt.figure(figsize=(20,10))
plt.plot(K, distortions, color = 'g')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.colormaps();
plt.show()
```

```python
# run Kmeans clustering algorithm
data1=pd.read_csv('Dataset2.csv')
k=4
kmeans = KMeans(n_clusters=k, random_state=0).fit(data1)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
plt.scatter(data1['X'],data1['Y'], c=kmeans.labels_, cmap='rainbow')
plt.show()
```
